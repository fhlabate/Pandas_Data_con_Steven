Pandas tiene dos estructuras de datos principales, son:

* **DataFrame**: Similar a una hoja de cálculo en Excel o una tabla en una base de datos, un DataFrame de Pandas es una **estructura bidimensional** que organiza los datos en filas y columnas. Cada columna puede contener diferentes tipos de datos (números, texto, fechas, etc.).

* **Series**: Es una **estructura unidimensional** similar a una lista o un vector, pero con etiquetas en lugar de índices numéricos. Puedes pensar en una Serie como una sola columna de un DataFrame

## [Tipos de Datos](https://www.youtube.com/watch?v=p3DJotm82d4&list=PLEiS2zOIb3ric1EU1l6XJ8iOKxClvqXht&index=3)

Tipos de datos en Pandas, tomando en cuenta la capacidad de soportar datos nulos:

1. **float:** Es el tipo de dato de punto flotante de NumPy y puede contener valores decimales. Soporta datos nulos (NaN).

2. **int:** Es el tipo de dato de número entero de NumPy y no soporta datos nulos.

3. **Int64:** Es un tipo de dato introducido por Pandas que representa números enteros y puede manejar datos nulos. Es especialmente útil cuando necesitas representar números enteros con valores nulos en tu DataFrame.

4. **object:** Es el tipo de dato de objeto de NumPy y se utiliza para representar texto o cualquier objeto de Python. Es versátil pero no es la elección óptima cuando se trata de datos numéricos o categóricos.

5. **category:** Es un tipo de dato introducido por Pandas específicamente diseñado para variables categóricas. Puede ayudar a ahorrar memoria y acelerar ciertas operaciones cuando trabajas con un conjunto de categorías limitado.

6. **bool:** Es el tipo de dato booleano de NumPy y no soporta datos nulos.

7. **boolean:** Es un tipo de dato booleano introducido por Pandas que puede manejar datos nulos. Esto es útil cuando necesitas representar valores booleanos con valores nulos.

8. **datetime64:** Es el tipo de dato de fecha y hora de NumPy y no soporta datos nulos.

```python
#Con el método ZIP
diccionario = dict(zip(indice,valores))
diccionario
```

```python
serie_dict = pd.Series(diccionario)
serie_dict
```

### Crear un Dataframe

#### Desde un diccionario

```python
dicc1 = {'nombre':['Pedro','Maria','Juan'],'edad': [30,40,50]}
print ('Este es el diccionario: \n\n',pd.Series(dicc1))

df_dicc = pd.DataFrame(dicc1,['10','20','30']) #No es necesario poner el índice. Es a modo lectivo
print('\n\nEste es el diccionario en DataFrame: ',df_dicc)
```

## [Indexación - Parte 2](https://www.youtube.com/watch?v=BII8t8So48I&list=PLEiS2zOIb3ric1EU1l6XJ8iOKxClvqXht&index=7)

### Cambiar de Multiindice a Multicolumns

El método .unstack() en pandas se utiliza comúnmente en situaciones donde los datos están en un formato apilado o en un índice jerárquico (MultiIndex) y quieres descomponer una de las capas del índice jerárquico en columnas separadas para facilitar el análisis y la visualización de los datos.

```python
clientes
```

```python
clientes_columns = clientes.unstack(fill_value=0) #Le pongo fill_value=0 para que no salgan NaN
clientes_columns
```

### De Multicolumns A Multiindice

```python
clientes_multindice = clientes_columns.stack()
clientes_multindice
```

```python
# Si lo quiero pasar solo a indice
clientes_multindice.stack().droplevel(-2)
clientes_multindice
```

### Como Indexar con Pandas

#### Indexar columnas
Hay una convención que es llamar 'df' al DataFrame que estamos analizando, ahora lo hemos estado llamando 'datos', pero lo vamos a pasar a 'df'

```python
df = pd.read_csv('Datos/Frutos_Secos.csv', delimiter=';', index_col='id')
df
```

#### Indexar por nombre

```python
# 1 columna
df.loc[:,'Nombre Cliente'] #[:,] para todas las filas
```

```python
#Varias columnas. ¡Atención al []!
df.loc[:,'Nombre Cliente':'Fecha Venta'] #.loc() es la ventaja si sabes el nombre de las columnas
```

#### Por posición

```python
df
```

```python
df.iloc[:,0] #"i"loc por "indice"
```

```python
#Primera y segunda columna
df.iloc[:,0:2] #"i"loc por "indice"
```

#### Por criterio

```python
#Las columnas que contenga "Cliente" en su nombre
df.loc[:,df.columns.str.contains('Cliente')]
```

```python
#Que contenga la palabra "Cliente" y que tambien me ponga aquellas que empiezan por "Fecha"
df.loc[:,df.columns.str.contains('Cliente') | df.columns.str.startswith('Fecha')]
```
